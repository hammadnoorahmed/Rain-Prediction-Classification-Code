```python
# working to remove the outliers from all numerical columns
```

```python
# Assuming df is your dataframe

# Select only numerical columns
numerical_df = df.select_dtypes(include=['float', 'int'])

# Calculate the quartiles
Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[~((numerical_df < lower_bound) | (numerical_df > upper_bound)).any(axis=1)]
```

```python
Y= updated_df['RainTomorrow'].values
```

```python
X = updated_df.drop(columns=['Date', 'RainTomorrow']).values

```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
updated_df.info()
```

```python
X[:, 0] = labelencoder.fit_transform(X[:, 0])

X[:, 4] = labelencoder.fit_transform(X[:, 4])
X[:, 6] = labelencoder.fit_transform(X[:, 6])
X[:, 7] = labelencoder.fit_transform(X[:, 7])
X[:, 16] = labelencoder.fit_transform(X[:, 16])
```

```python
X
```

```python
X1 = pd.DataFrame(X)
X1
```

```python
onehot_encoder = OneHotEncoder(categories='auto', sparse=False)
```

```python
columns_to_encode = [0, 4, 6, 7, 17, 18] #we will not one hot code for rain today because it is already only have two values
for i in columns_to_encode:
    data = X1[i].values.reshape(-1, 1)
    
    # Fit and transform the selected column with one-hot encoding
    X_encoded = onehot_encoder.fit_transform(data)
    encoded_column_names = onehot_encoder.get_feature_names_out(input_features=[str(i)])
    
    # Convert the transformed column back to DataFrame
    encoded_df = pd.DataFrame(X_encoded, columns=encoded_column_names)
    
    # Drop the original column from X1
    X1.drop(columns=i, inplace=True)
    
    # Concatenate the encoded column to X1
    X1 = pd.concat([X1, encoded_df], axis=1)
```

```python
X = X1.values
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=42, test_size=0.2)
```

Entropy using

```python
clf_entropy= DecisionTreeClassifier(criterion='entropy', random_state=100, 
                                   max_depth=5, min_samples_leaf=8)
```

```python
clf_entropy.fit(X_train, Y_train)
```

```python
Y_pred = clf_entropy.predict(X_test)
Y_pred
```

```python
print('Accuracy is'), accuracy_score(Y_test, Y_pred)*100
```

Using Gini

```python
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_split=3, min_samples_leaf=6, max_features=None, random_state=42)
```

```python
Y_pred = clf_gini.predict(X_test)
Y_pred
```

```python
print('Accuracy is'), accuracy_score(Y_test, Y_pred)*100
```
